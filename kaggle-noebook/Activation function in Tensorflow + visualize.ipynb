In this section, we import the essential libraries for our project:

```python
import tensorflow as tf
from tensorflow.keras.utils import plot_model
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set_theme(style="whitegrid", context="talk", font_scale=1)
```

## üé® Visualizing Activation Functions

This function, `draw_activation`, is designed to visualize different activation functions in TensorFlow. Here's a breakdown of its components:

- **Parameters**:
  - `name`: The name of the activation function to visualize (default: `'relu'`).
  - `start` and `end`: Define the range of input values for the function (default: `-10` to `10`).

- **Process**:
  1. The function creates an activation layer using `tf.keras.layers.Activation`.
  2. It generates input values using `np.linspace` for a smooth curve.
  3. Applies the activation function to the inputs and plots the results using `matplotlib`.

- **Error Handling**:
  - If the specified activation function is invalid, the function gracefully prints `"Error"`. üõë

Feel free to test this function with different activation names like `sigmoid`, `tanh`, or `linear`! üéâ


```python
def draw_activation(name='relu', start=-10, end=10):
    try:
        f = tf.keras.layers.Activation(name)
        x = np.linspace(start, end, 500)
        y = f(x)
        plt.figure(figsize=(15, 6))
        plt.plot(x, y)
        plt.title(name)
    except Exception:
        print("Error")
        
```

## üîç Activation Function: ReLU (Rectified Linear Unit)

### üßÆ Formula:
The ReLU activation function is defined as:

$ f(x) = \max(0, x) $

This means it outputs the input directly if it is positive; otherwise, it returns 0.

### üí° Usage:
- ReLU is one of the most commonly used activation functions in deep learning, particularly in hidden layers of neural networks. 
- It's widely used in Convolutional Neural Networks (CNNs) and deep feedforward networks. üåü

### üìà Applications:
- Image classification tasks.
- Natural language processing models.
- Deep reinforcement learning networks.

### ‚öôÔ∏è Performance:
- ReLU introduces non-linearity into the model, which is crucial for learning complex patterns in data. 
- Its simple formula makes it computationally efficient.

### ‚úÖ Advantages:
1. **Efficiency**: Faster computation due to its simplicity.
2. **Sparse Activation**: Only positive neurons activate, which helps reduce computational load.
3. **Vanishing Gradient Problem Mitigation**: Compared to sigmoid or tanh, ReLU helps in avoiding vanishing gradient issues.

### ‚ùå Disadvantages:
1. **Dying ReLU Problem**: Neurons can sometimes get stuck outputting zero for all inputs, leading to no learning.
2. **Not Symmetric**: Outputs are asymmetric for positive and negative inputs, which might affect certain applications.

### üõ†Ô∏è Tips:
- To avoid the "dying ReLU" problem, consider using variants like **Leaky ReLU** or **Parametric ReLU (PReLU)**.

## üîç Activation Function: ELU (Exponential Linear Unit)

### üßÆ Formula:
The ELU activation function is defined as:

$$
f(x) =
\begin{cases} 
x & \text{if } x > 0 \\
\alpha (e^x - 1) & \text{if } x \leq 0
\end{cases}
$$

Where $\alpha$ is a parameter that determines the value for negative inputs (commonly set to 1).

### üí° Usage:
- ELU is often used in deep learning models as an alternative to ReLU to address its limitations.
- It introduces smoother transitions for negative values, improving gradient flow during training.

### üìà Applications:
- Deep neural networks where robust gradient flow is critical.
- Tasks requiring smooth and continuous activation functions.

### ‚öôÔ∏è Performance:
- ELU helps avoid the "dying ReLU" problem by providing non-zero outputs for negative inputs.
- It is computationally more intensive than ReLU but offers better performance in certain scenarios.

### ‚úÖ Advantages:
1. **Gradient Flow**: Provides smoother gradients for inputs close to zero.
2. **Avoids Dying Neurons**: Outputs are non-zero for negative inputs.
3. **Continuity**: The function is smooth and differentiable, which helps in stable optimization.

### ‚ùå Disadvantages:
1. **Computational Overhead**: Slightly more expensive than ReLU due to the exponential calculation.
2. **Parameter Dependency**: Performance depends on the choice of $\alpha$, which might require tuning.

### üõ†Ô∏è Tips:
- Use ELU when training deep networks where ReLU might suffer from "dying neurons."
- Experiment with different $\alpha$ values to optimize performance.


```python
draw_activation('elu')
```

## üîç Activation Function: SELU (Scaled Exponential Linear Unit)

### üßÆ Formula:
The SELU activation function is defined as:

$$
f(x) =
\begin{cases} 
\lambda x & \text{if } x > 0 \\
\lambda \alpha (e^x - 1) & \text{if } x \leq 0
\end{cases}
$$

Where:
- $\lambda \approx 1.0507$ (a scaling factor to ensure normalized outputs).
- $\alpha \approx 1.6733$ (a parameter for negative inputs).

### üí° Usage:
- SELU is primarily used in Self-Normalizing Neural Networks (SNNs).
- It ensures that the mean and variance of activations remain normalized during training.

### üìà Applications:
- Deep learning tasks where internal normalization is essential.
- Often used in architectures with a lot of layers to maintain stable training.

### ‚öôÔ∏è Performance:
- SELU enables self-normalizing properties by keeping activations close to zero mean and unit variance.
- It reduces the need for techniques like batch normalization.

### ‚úÖ Advantages:
1. **Self-Normalizing**: Helps maintain normalized outputs, leading to faster convergence.
2. **No Vanishing Gradient**: Activates consistently, even for negative inputs.
3. **Improves Stability**: Reduces the chances of exploding or vanishing gradients.

### ‚ùå Disadvantages:
1. **Input Constraints**: Works best when inputs are normalized (mean of 0, variance of 1).
2. **Parameter Dependency**: Requires specific $\lambda$ and $\alpha$ values to achieve its benefits.

### üõ†Ô∏è Tips:
- Normalize your inputs before using SELU to achieve the best results.
- Use SELU with architectures designed for Self-Normalizing Neural Networks.


```python
draw_activation('selu')
```

## üîç Activation Function: Swish

### üßÆ Formula:
The Swish activation function is defined as:

$$
f(x) = x \cdot \text{sigmoid}(x)
$$

Where $\text{sigmoid}(x)$ is the standard sigmoid function:

$$
\text{sigmoid}(x) = \frac{1}{1 + e^{-x}}
$$

This means Swish scales the input $x$ by the sigmoid of $x$.

### üí° Usage:
- Swish is a smooth and non-monotonic activation function introduced by researchers at Google.
- It has been shown to outperform ReLU and other traditional activation functions in some deep learning models.

### üìà Applications:
- Widely used in image classification and deep neural networks.
- Suitable for tasks where smooth gradients improve optimization.

### ‚öôÔ∏è Performance:
- Swish allows small negative values, which can help improve gradient flow compared to ReLU.
- It adapts dynamically to inputs, making it more flexible for complex models.

### ‚úÖ Advantages:
1. **Smoothness**: Provides a smoother gradient transition compared to ReLU.
2. **Improved Gradient Flow**: Handles small negative values better, reducing the "dying neuron" problem.
3. **Dynamic Range**: Automatically adjusts based on input magnitude.

### ‚ùå Disadvantages:
1. **Computational Overhead**: Slightly more expensive than ReLU due to the sigmoid computation.
2. **Non-Monotonicity**: May not be ideal for simpler models where monotonic functions perform better.

### üõ†Ô∏è Tips:
- Use Swish for deep networks requiring smooth activation functions and improved gradient flow.
- It's particularly effective in architectures like EfficientNet where performance gains are needed.


```python
draw_activation('swish')
```

## üîç Activation Function: Sigmoid

### üßÆ Formula:
The Sigmoid activation function is defined as:

$$
f(x) = \frac{1}{1 + e^{-x}}
$$

It maps the input $x$ to a value between 0 and 1, making it a smooth and differentiable function.

### üí° Usage:
- The Sigmoid function is commonly used in binary classification problems and as an activation function in the output layer of neural networks for probabilities.

### üìà Applications:
- Logistic regression models.
- Output layers of binary classification neural networks.
- Situations where outputs need to represent probabilities.

### ‚öôÔ∏è Performance:
- The sigmoid function compresses a wide range of input values into the interval (0, 1), but it can suffer from the vanishing gradient problem for large positive or negative inputs.

### ‚úÖ Advantages:
1. **Probability Interpretation**: Outputs are bounded between 0 and 1, making it suitable for probability-based predictions.
2. **Smooth Gradient**: Provides a smooth gradient, which can aid optimization.

### ‚ùå Disadvantages:
1. **Vanishing Gradient Problem**: Gradients become very small for large positive or negative inputs, slowing learning.
2. **Non-Zero Centered**: Outputs are not zero-centered, which can make gradient-based optimization less efficient.

### üõ†Ô∏è Tips:
- Use Sigmoid primarily in the output layer for binary classification tasks.
- Avoid using Sigmoid in hidden layers of deep networks; prefer ReLU or its variants to mitigate vanishing gradient issues.


```python
draw_activation('sigmoid')
```

## üîç Activation Function: Tanh (Hyperbolic Tangent)

### üßÆ Formula:
The Tanh activation function is defined as:

$$
f(x) = \tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}
$$

It maps the input $x$ to a value between $-1$ and $1$, making it symmetric around zero.

### üí° Usage:
- The Tanh function is commonly used in hidden layers of neural networks to handle both positive and negative values.
- It is preferred over Sigmoid when the mean of the data is centered around zero.

### üìà Applications:
- Deep learning tasks where inputs are normalized or centered.
- Hidden layers of feedforward and recurrent neural networks.

### ‚öôÔ∏è Performance:
- Tanh provides stronger gradients than Sigmoid for inputs near zero, leading to faster convergence during training.
- However, it suffers from the vanishing gradient problem for large positive or negative inputs.

### ‚úÖ Advantages:
1. **Symmetry**: Outputs are zero-centered, making gradient-based optimization more efficient.
2. **Smooth Gradient**: Provides smooth gradients that are stronger than Sigmoid near zero.

### ‚ùå Disadvantages:
1. **Vanishing Gradient Problem**: Gradients diminish for large inputs, slowing learning.
2. **Computational Cost**: Slightly more expensive to compute than ReLU.

### üõ†Ô∏è Tips:
- Use Tanh in hidden layers when the input data is normalized to have zero mean.
- For deeper networks, consider ReLU or its variants to avoid the vanishing gradient problem.


```python
draw_activation('tanh')
```

## üîç Activation Function: Linear

### üßÆ Formula:
The Linear activation function is defined as:

$$
f(x) = x
$$

This means the output is directly equal to the input without any transformation.

### üí° Usage:
- The Linear activation function is often used in the output layer of neural networks for regression tasks, where the output is not bounded and can take any value.

### üìà Applications:
- Regression problems where the target variable is continuous.
- Neural networks requiring unbounded output values.
- Can act as a "pass-through" layer when no activation is needed.

### ‚öôÔ∏è Performance:
- Linear activation does not introduce non-linearity into the model, which limits the network's ability to model complex patterns. 
- It is computationally efficient since no additional transformation is applied.

### ‚úÖ Advantages:
1. **Simplicity**: Easy to compute with no additional overhead.
2. **Unbounded Output**: Suitable for regression tasks where predictions are not limited to a specific range.
3. **No Vanishing Gradient**: Gradients do not diminish, ensuring stable learning.

### ‚ùå Disadvantages:
1. **No Non-Linearity**: Cannot model complex relationships in data, limiting its usefulness in hidden layers.
2. **Limited Use**: Primarily used in output layers of regression tasks, not hidden layers.

### üõ†Ô∏è Tips:
- Use Linear activation in the output layer for regression tasks or when unbounded outputs are required.
- Avoid using Linear in hidden layers as it limits the model's ability to learn non-linear patterns.


```python
draw_activation('linear')
```
